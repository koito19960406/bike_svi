```python
import os
import pandas as pd
import geopandas as gpd
```

```python
population = pd.read_excel("/Volumes/ExFAT/bike_svi/data/external/cities/London/control_variables/population/mid-2011-lsoa-quinary-estimates.xls", sheet_name="Mid-2011 Persons", header=None).iloc[4:,:]
population
```

```python
population.columns = population.iloc[0,:]
# remove the 1st row
population = population.iloc[1:,:]
population = population.drop(population.iloc[:,1:4], axis =1)
```

```python
population
```

```python
age_counter = 0
while age_counter <= 80:
    print(population)
    # if year != 2011:
    #     # sum up 10 columns at a time
    population[f"{str(age_counter)} - {str(age_counter+9)}"] = population.iloc[:,1:3].sum(axis=1)
    
    # drop columns
    population.drop(population.iloc[:,1:3], inplace=True, axis=1)
    # update age_counter
    age_counter+=10
    
    if age_counter == 90:
        population["90+"] = population.iloc[:,1]
        population.rename(columns={population.columns[0]: "lsoa_code", population.columns[1]: "90+"}, inplace=True)
```
